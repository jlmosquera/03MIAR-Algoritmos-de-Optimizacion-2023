## Descenso del Gradiente - AGD. Pr√°ctica

## 1. Usando una tasa de aprendizaje constante

```python
## 1. Importar la librerias necesarias
import math                           # Funciones matem√°ticas (definidas en C)
import matplotlib.pyplot as plt       # Creaci√≥n de gr√°ficos en dos dimensiones
import numpy as np                    # Calculo num√©rico y an√°lisis de datos
import random                         # Generadores de n√∫meros pseudo-aleatorios

f = lambda X: X[0] ** 2 + X[1] ** 2   # Funci√≥n
df = lambda x: [2*x[0], 2*x[1]]       # Gradiente

## 2. Preparar de los datos para dibujar mapa de niveles de Z
resolucion = 30
rango = 2.5
X = np.linspace(-rango, rango,  resolucion)
Y = np.linspace(-rango, rango, resolucion)
Z = np.zeros((resolucion, resolucion))

for ix, x in enumerate(X):
  for iy, y in enumerate(Y):
    Z[ix, iy] = f([x, y])

## 3. Pintar el mapa
contour = plt.contourf(X, Y, Z, resolucion)
cbar = plt.colorbar(contour)

## 4. Generar del punto de partida aleatorio
random.seed(20230619)         # Inicializaciomos la raiz para sureproducibilidad
P = [random.uniform(-rango, rango), random.uniform(-rango, rango)]
plt.plot(P[0], P[1], "o", c = "red")
TA = 0.05

## 5. Buscar el m√≠nimo de la funci√≥n
for _ in range(500):
  grad = df(P)
  P[0], P[1] = P[0] - TA * grad[0], P[1] - TA * grad[1]
  plt.plot(P[0], P[1], "o", c = "red")

## Pintar el grafico con las interacciones
plt.plot(P[0], P[1], "o", c = "white")
plt.show()
print("Solucion:", P, f(P))
```
